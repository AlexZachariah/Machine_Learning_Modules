```python
def summaryOfData(df):
    print('1.The shape of the input data is:',df.shape)
    print('='*60+'\n')
    print('2.The size of the input data is:',df.size)
    print('='*60+'\n')
    print('3.dtype of the data is:',df.dtypes)
    print('='*60+'\n')
    print('4.info about the data is:','\n')
    print(df.info())
    print('='*60+'\n')
    print('5.data is described as:',df.describe())
    print('='*60+'\n')
    print('6.head of data is:',df.head())
    print('='*60+'\n')
    print('7.tail of data is:',df.tail())
    print('='*60+'\n')
    print('8.null value od data:',df.isnull())
    print('='*60+'\n')
    print('9.sum of null value of data is:',df.isnull().sum())
```

```python
df_customer =pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Tom', 'Jenny', 'James', 'Dan']
})

df_info = pd.DataFrame({
    'id': [2, 3, 4, 5],
    'age': [31, 20, 40, 70],
    'sex': ['F', 'M', 'M', 'F']
})
```

```python
pd.merge(df_customer,df_info)
```

```python
df_customer.merge(df_info)
```

```python
pd.merge(df_customer,df_info,on='id')
```

```python
df_order = pd.DataFrame({
    'id': [2,3,4,5],
    'name': ['Jenny', 'James', 'Dan', 'leo'],
    'quantity': [2,4,6,10]
})
df_order
```

```python
pd.merge(df_customer,df_order)
```

```python
df_customer = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['Tom', 'Jenny', 'James', 'Dan'],
})
df_info = pd.DataFrame({
    'id': [2,3,4,5],
    'age': [31,20,40,70],
    'sex': ['F', 'M', 'M', 'F']
})
```

```python
df_customer
```

```python
df_info
```

```python
#inner join or normal

pd.merge(df_customer,df_info,on='id')
```

```python
#left join

pd.merge(df_customer,df_info,on='id',how='left')
```

```python
#right join

pd.merge(df_customer,df_info,on='id',how='right')
```

```python
#outer join

pd.merge(df_customer,df_info,on='id',how='outer')
```
