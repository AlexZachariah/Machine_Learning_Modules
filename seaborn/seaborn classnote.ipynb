```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
## Count plot

sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot('time',data=df)
```

```python
## Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
## Bar plot
sns.barplot('sex','total_bill',data=df)
```

```python
sns.barplot('day','total_bill',data=df)
```

```python
#boxplot

sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('total_bill', data=df)
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.boxplot('day','tip',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
sns.boxplot(x="total_bill", y="day", hue="sex",data=df)
```

```python
#violinplot

sns.violinplot(x="total_bill", y="day", data=df)
```

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
